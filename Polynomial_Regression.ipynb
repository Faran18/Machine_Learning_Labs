```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv('/content/cars.csv')

dataset = dataset.drop('car_names', axis=1)

X = dataset.drop('miles_per_gallon', axis=1)
y = dataset['miles_per_gallon']
```

#Multi regression tasks

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.4, random_state = 42)
X_train.shape
```

```python
multiple_regressor = LinearRegression()
multiple_regressor.fit(X_train, y_train)

print( multiple_regressor.coef_)
print( multiple_regressor.intercept_)

```

```python
y_pred_multiple = multiple_regressor.predict(X_test)
```

```python
print(f'\nMultiple Regression - Variance score (R^2): {multiple_regressor.score(X_test, y_test)}')
print('MAE:', metrics.mean_absolute_error(y_test, y_pred_multiple))
print('MSE:', metrics.mean_squared_error(y_test, y_pred_multiple))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_multiple)))
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(multiple_regressor.predict(X_train), multiple_regressor.predict(X_train) - y_train, color="green", s=10, label='Train data')
plt.scatter(multiple_regressor.predict(X_test), multiple_regressor.predict(X_test) - y_test, color="blue", s=10, label='Test data')
plt.hlines(y=0, xmin=0, xmax=50, linewidth=2)
plt.legend(loc='upper right')
plt.title("Residual errors for Multiple Regression (40% test data)")
plt.xlabel("Predicted Miles_Per_Gallon")
plt.ylabel("Error")
plt.show()
```

```python
feature = 'quarter_mile _time'
print(f'\nPerforming Polynomial Regression on {feature.capitalize()}:')

X_polynomial = X[[feature]].values

for degree in [2, 3, 4]:
    print(f'\nAnalyzing {degree}-Degree Polynomial Regression for {feature.capitalize()}:')


    polynomial_features = PolynomialFeatures(degree=degree)
    X_polynomial_train = polynomial_features.fit_transform(X_train[[feature]])
    X_polynomial_test = polynomial_features.transform(X_test[[feature]])

    polynomial_model = LinearRegression()
    polynomial_model.fit(X_polynomial_train, y_train)

    print(f"Model Coefficients: {polynomial_model.coef_}")
    print(f"Model Intercept: {polynomial_model.intercept_}")


    y_predicted_polynomial = polynomial_model.predict(X_polynomial_test)


    print(f'Polynomial Regression (Deg {degree}) - R² Score: {polynomial_model.score(X_polynomial_test, y_test):.4f}')
    print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_test, y_predicted_polynomial))
    print('Mean Squared Error (MSE):', metrics.mean_squared_error(y_test, y_predicted_polynomial))
    print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, y_predicted_polynomial)))


    plt.figure(figsize=(12, 8))
    plt.scatter(X_train[feature], y_train, color='red', label='Training Data')
    plt.plot(np.sort(X_train[feature]),
             polynomial_model.predict(polynomial_features.transform(np.sort(X_train[feature]).reshape(-1, 1))),
             color='blue', label=f'{degree}-Degree Fit')
    plt.title(f'{degree}-Degree Polynomial Regression for {feature.capitalize()}')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Miles_per_Gallon (MPG)')
    plt.legend()
    plt.show()
```
